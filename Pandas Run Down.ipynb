# Grouping and Aggregation 

```python
#grouping using mtcars data set, import data and assign to cars
file_path = "mtcars.csv"
cars = pd.read_csv(file_path)
cars.head(10)
```

```python
#assign column names  ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb'] using columns att
cars.columns = ['car_names','mpg','cyl','disp','hp','drat','wt','qsec','vs','am','gear','carb']
cars.head()
```

```python
#grouping cars_groups = cars.groupby(cars['cyl'])
cars_groups = cars.groupby(cars['cyl'])
group_report = cars_groups.mean()
```

```python
group_report
```

```python
writer = pd.ExcelWriter('report.xlsx')
group_report.to_excel(writer,'Sheet1',index=False)
writer.save()
```
